```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import VarianceThreshold
```

```python
data = pd.read_csv('/Users/mac/Desktop/菜菜的sklearn课堂课件/03数据预处理和特征工程/digit recognizor.csv')
data.head()
```

```python
X = data.iloc[:, 1:]
y = data.iloc[: ,0]
```

```python
X.shape
```

```python
# 去掉 方差<=0 的features
X_var0 = VarianceThreshold(threshold=0).fit_transform(X)
pd.DataFrame(X_var0).head()
```

```python
# 保留一半的 features
X_fsvar = VarianceThreshold(threshold=np.median(X.var(axis=0).values)).fit_transform(X)
pd.DataFrame(X_fsvar).head()
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC  # 使用随机森林是因为随机森林很快
from sklearn.model_selection import cross_val_score
from sklearn.feature_selection import SelectKBest, chi2 # 卡方检验chi2
```

```python
# 卡方检验

# 假设我们需要 300个 features
X_fschi = SelectKBest(chi2, k=300).fit_transform(X_fsvar, y)
X_fschi.shape
```

```python
cross_val_score(RFC(n_estimators=10, random_state=0), X_fschi, y, cv=5).mean()
```

```python
# hyperparameter k 的选取
chivalue, pvalues_chi = chi2(X_fsvar, y)
```

```python
pvalues_chi
```

```python
# 消除p值大于设定值的所有特征
k = chivalue.shape[0] - (pvalues_chi > 0.05).sum()
k
```

```python
# F检验

from sklearn.feature_selection import f_classif
```

```python
F, pvalues_f = f_classif(X_fsvar, y)
```

```python
k = F.shape[0] - (pvalues_f > 0.05).sum()
```

```python
k
```

```python
# 互信息法

from sklearn.feature_selection import mutual_info_classif as MIC
```

```python
result = MIC(X_fsvar, y)
```

```python
result
```

```python
k = result.shape[0] - sum(result <= 0)
k
```

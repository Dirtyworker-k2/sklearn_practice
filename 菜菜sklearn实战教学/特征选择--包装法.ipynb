```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.feature_selection import RFE
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
data = pd.read_csv('/Users/mac/Desktop/菜菜的sklearn课堂课件/03数据预处理和特征工程/digit recognizor.csv')
X = data.iloc[:, 1:]
y = data.iloc[:, 0]
```

```python
RFC_ = RFC(n_estimators=10, random_state=0)
```

```python
X.shape
```

```python
# Features selector -- wrapper method(RFE)
selector = RFE(RFC_, estimator=RFC_, n_features_to_select=340, step=50).fit(X, y)
```

```python
selector.support_.sum()
```

```python
selector.ranking_
```

```python
# 参数中无y
X_wrapper = selector.transform(X)
```

```python
cross_val_score(RFC_, X_wrapper, y, cv=5).mean()
```

```python
X.loc[:, selector.support_]
```

```python
# choice of hyperparameter - n_features_to_select
score = [] 
for i in range(1, 751, 50):
    X_wrapper = RFE(RFC_, n_features_to_select=i, step=50).fit_transform(X, y)
    score.append(cross_val_score(RFC_, X_wrapper, y, cv=5).mean())
```

```python
fig = plt.figure(figsize=(18, 6))
plt.plot([*range(1, 751, 50)], score)
plt.xticks([*range(1, 751, 50)])
plt.show()
```

```python
from sklearn import tree 
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
# 加载 sklearn.load_wine 的数据集
wine = load_wine()
```

```python
wine.data
```

```python
wine.feature_names
```

```python
import pandas as pd 
```

```python
wine.target
```

```python
# 将 wine 的特征矩阵转换成 DataFrame 形式
feature = pd.DataFrame(wine.data, columns=wine.feature_names)
```

```python
feature
```

```python
# 将 wine 的标签转换成 DataFrame 形式
target = pd.DataFrame(wine.target, columns=['target'])
```

```python
# 将两个DataFrame组合在一起，形成新的DataFrame
pd.concat([feature, target], axis=1)
```

```python
# 将数据集分开，分为训练集和测试集

X_train, X_test, y_train, y_test = train_test_split(wine.data, wine.target, test_size=0.3, random_state=0)
```

```python
len(wine.feature_names)
```

```python
# 决策树模型-- 1、实例化
clf = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
# 决策树模型-- 2、训练
clf.fit(X_train, y_train)
```

```python
# 决策树模型-- 3、打分
clf.score(X_test, y_test)
```

```python
# 决策树可视化

import graphviz
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类','花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']
```

```python
dot_data = tree.export_graphviz(clf
                                ,feature_names = feature_name
                                ,class_names = ['琴酒', '雪梨', '贝尔摩德']
                                ,filled=True
                                ,rounded=True
                                ,out_file=None)
```

```python
graph = graphviz.Source(dot_data)
```

```python
graph
```

```python
# 查看哪些features对决策树的性能有巨大的影响
clf.feature_importances_
```

```python
# 将feature_names 与 feature_importances_ 一一对应
[*zip(wine.feature_names, clf.feature_importances_)]
```

```python
# splitter -- 涉及节点选择； random_state -- 涉及节点选择

clf = tree.DecisionTreeClassifier(criterion='entropy'
                                 ,splitter='random'
                                 ,random_state=30)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
```

```python
# 涉及一些剪枝参数: max_depth、min_samples_leaf、min_samples_split

clf = tree.DecisionTreeClassifier(criterion='entropy'
                                 ,splitter='random'
                                 ,random_state=30
                                 ,max_depth=3
                                 ,min_samples_leaf=10
                                 ,min_samples_split=10)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
dot_data = tree.export_graphviz(clf
                               ,feature_names=wine.feature_names
                               ,class_names=['琴酒', '雪梨', '贝尔摩德']
                               ,filled=True
                               ,rounded=True
                               ,out_file=None)
graph = graphviz.Source(dot_data)
graph
```

```python
import matplotlib.pyplot as plt 
```

```python
# Learning Curves -- max_depth的学习曲线

test = []
for i in range(10):
    clf = tree.DecisionTreeClassifier(max_depth=i+1
                                     ,criterion='entropy'
                                     ,random_state=30)
    clf.fit(X_train, y_train)
    score = clf.score(X_test, y_test)
    test.append(score)

plt.plot(range(1, 11), test, color='red', label='max_depth')
plt.legend(loc=4)
plt.show()
```

```python
import pandas as pd
```

```python
# 从本地文件夹中读取 csv格式 的文件
# 可以用data.head() 、 data.info()查看DataFrame的具体信息。
# 用data.isnull().sum()查看空缺值情况

data = pd.read_csv('/Users/mac/desktop/未命名项目/test/data.csv')
data.head()
```

```python
data.info()
```

```python
# 1、筛选features，去掉无用的features

data.drop(['Name', 'Ticket', 'Cabin'], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
# 2、 空缺值处理
# 用平均值填充某feature的空缺
data['Age'] = data['Age'].fillna(data['Age'].mean())
```

```python
data.info()
```

```python
#  去掉含有空缺值的samples
data = data.dropna(axis=0)
```

```python
data.info()
```

```python
data.head(10)
```

```python
# 3、object --> number：机器学习只能处理数字

labels = data.loc[:, 'Embarked'].unique().tolist()  # pd.Series.unique(), pd.DataFrame没有unique()方法
```

```python
labels
```

```python
# list.index() 得到数组中某元素的指数

data.loc[:, 'Embarked'] = data.loc[:, 'Embarked'].apply(lambda x: labels.index(x))
```

```python
data.head(10)
```

```python
data.loc[:, 'Sex'] = (data.loc[:, 'Sex'] == 'male').astype('int')
```

```python
data.head()
```

```python
data.iloc[:, 3]
```

```python
data.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# 筛选出某一features
X = data.iloc[:, data.columns != 'Survived']
```

```python
X
```

```python
data.columns, data.index
```

```python
data.columns != 'Survived'
```

```python
y = data.iloc[:, data.columns == 'Survived']
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.head()
```

```python
# 使用train_test_split 后，特征矩阵的index会乱，可以重新使它们从小到大排序(单个)

X_train.index = range(X_train.shape[0])
```

```python
X_train.head(10)
```

```python
# 使用train_test_split 后，特征矩阵的index会乱，可以重新使它们从小到大排序（循环）

for dataset in X_train, X_test, y_train, y_test:
    dataset.index = range(dataset.shape[0])
```

```python
X_test.head(10)
```

```python
# 训练clf模型
```

```python
clf = DecisionTreeClassifier(random_state=25)
clf.fit(X_train, y_train)
score = clf.score(X_test, y_test)
```

```python
score
```

```python
clf = DecisionTreeClassifier(random_state=25)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score = cross_val_score(clf, X, y, cv=10, scoring='accuracy').mean()
```

```python
score
```

```python
# Learning Curves  超参数学习曲线
scores_tr = []
scores_te = []

for i in range(11):
    clf = DecisionTreeClassifier(random_state=25
                                ,max_depth=i+1
                                ,criterion='entropy')
    clf.fit(X_train, y_train)
    score_tr = clf.score(X_train, y_train)
    score_te = cross_val_score(clf, X, y, cv=10, scoring='accuracy').mean()
    scores_tr.append(score_tr)
    scores_te.append(score_te)
```

```python
import matplotlib.pyplot as plt 
```

```python
fig, ax = plt.subplots()
ax.plot(range(11), scores_tr, color='red', linewidth=2, label='Train Error')
ax.plot(range(11), scores_te, color='blue', linewidth=2, label='Test Error')
plt.legend(loc=2)
print(max(scores_te))
```

```python
# GridSearch 超参数选择  

from sklearn.model_selection import GridSearchCV
```

```python
clf = DecisionTreeClassifier(random_state=25)
```

```python
import numpy as np

parameters = {'criterion': ('gini', 'entropy')
             ,'splitter': ('best', 'random')
             ,'max_depth': [*range(1, 10)]
             ,'min_samples_leaf': [*range(1, 50, 5)]
             ,'min_impurity_decrease': np.linspace(0, 0.5, 50)
}

gs = GridSearchCV(clf, parameters, cv=10) # fit\score\cross_validation
gs = gs.fit(X_train, y_train)
```

```python
gs.best_params_
```

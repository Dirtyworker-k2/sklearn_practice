```python
import pandas as pd 
import numpy as np
```

```python
data = pd.read_csv('/Users/mac/Desktop/未命名项目/test/Narrativedata.csv', index_col=0)  #请把第0列作为索引
data.head()
```

```python
data.info()
```

```python
Age = data.loc[:, 'Age'].values.reshape(-1, 1)
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(strategy='mean')   # 用平均数填补
imp_median = SimpleImputer(strategy='median')   # 用中值填补
imp_0 = SimpleImputer(strategy='constant', fill_value=0)  # 用0填补
```

```python
# 填补缺失值，必须一列一列填补
imp_mean = imp_mean.fit_transform(Age)
imp_median = imp_median.fit_transform(Age)
imp_0 = imp_0.fit_transform(Age)
```

```python
imp_mean
```

```python
data.loc[:, 'Age'] = imp_mean
```

```python
data.info()
```

```python
Embark = data.loc[:, 'Embarked'].values.reshape(-1, 1)
```

```python
# strategy='most_frequent' 可以用于填写文字型的缺失值
imp_mode = SimpleImputer(strategy='most_frequent')
```

```python
imp_mode = imp_mode.fit_transform(Embark)
```

```python
data['Embarked'] = imp_mode
```

```python
data.info()
```

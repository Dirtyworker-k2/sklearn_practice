```python
from sklearn.model_selection import cross_val_score
from sklearn.datasets import load_boston
from sklearn.tree import DecisionTreeRegressor
```

```python
boston = load_boston()
reg = DecisionTreeRegressor(random_state=0)
```

```python
# 衡量决策数回归模型质量的指标是mean_squared_error(mse), neg_mean_squared_error = - mean_squared_error

cross_val_score(reg, boston.data, boston.target, cv=10, scoring='neg_mean_squared_error')
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeRegressor
```

```python
rng = np.random.RandomState(1)
target = 5 * rng.rand(80, 1)
```

```python
X = np.sort(target, axis=0)
y = np.sin(X).ravel()
```

```python
fig, ax = plt.subplots()
ax.scatter(X, y, s=20, edgecolors='black', c='orange')
```

```python
# 给数据集添加噪声

y[::5] += 3 * (0.5 - rng.rand(16))
```

```python
fig, ax = plt.subplots()
ax.scatter(X, y, s=20, edgecolors='black', c='orange')
```

```python
reg_1 = DecisionTreeRegressor(max_depth=2)
reg_2 = DecisionTreeRegressor(max_depth=5)
reg_1.fit(X, y)
reg_2.fit(X, y)
```

```python
# sklearn模型只接受 '多维' 的特征矩阵， 因此需要给特征矩阵升维--[:, np.newaixs]
# 生成训练集 -- 用于画图
X_test = np.arange(0, 5, 0.01)[:, np.newaxis]
```

```python
X_test
```

```python
y_1 = reg_1.predict(X_test)
y_2 = reg_2.predict(X_test)
```

```python
y_1
```

```python
fig, ax = plt.subplots()
ax.scatter(X, y, s=20, edgecolors='black', c='orange', label='data')
ax.plot(X_test, y_1, color='red', linewidth=2, label='max_depth=2')
ax.plot(X_test, y_2, color='yellowgreen', label='max_depth=5', linewidth=2)
ax.set_xlabel()
ax.set_ylabel()
ax.title('Decision Tree Regression')
plt.legend()
plt.show()
```
